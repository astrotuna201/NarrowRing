```python
%matplotlib inline
import numpy as np
import matplotlib 
import matplotlib.pyplot as plt
def f(i,m,N,j):
    return np.cos(float(m)*np.pi*float(i+j)/float(N))*np.cos(float(m)*np.pi*float(i-j)/float(N))
    #return float(m)*np.pi*float(i+j)/float(N)
plt.figure()
normalie=0.0015
# 600 particles
N=600.
i=1.
result=[]
for m in range(1,100):
    result.append(np.array([f(i,m,N,j) for j in range(1,int(N+1))]).sum())
normalizcon=result[0]
print(normalizcon)
resultr=[(-i+normalizcon)*N*normalie for i in result]
plt.plot(range(1,100),resultr,label='600 particles')
# 200 particles    
N=200.
i=1.
result=[]
for m in range(1,100):
    result.append(np.array([f(i,m,N,j) for j in range(1,int(N+1))]).sum())
normalizcon=result[0]
print(normalizcon)
result1=[(-i+normalizcon)*N*normalie for i in result]
plt.plot(range(1,100),result1,label='200 particles')
# 60 particles    
N=60.
i=1.
result=[]
for m in range(1,100):
    result.append(np.array([f(i,m,N,j) for j in range(1,int(N+1))]).sum())
normalizcon=result[0]
print(normalizcon)
result2=[(-i+normalizcon)*N*normalie for i in result]
plt.plot(range(1,100),result2,label='60 particles')
# 10000 particles
# 60 particles    
N=10000.
i=1.
result=[]
for m in range(1,100):
    result.append(np.array([f(i,m,N,j) for j in range(1,int(N+1))]).sum())
normalizcon=result[0]
print(normalizcon)
result3=[(-i+normalizcon)*N*normalie for i in result]
plt.plot(range(1,100),result3,label='10000 particles')

plt.ylabel('precession')
plt.xlabel('mode')
plt.legend()


#print(298.98357822497087/98.95090275639029)
#print(98.95090275639029/28.83839591336406)
#print(600./200.)
#print(200./60.)
```

```python
# percentage errors
plt.figure()
plt.plot(range(2,100),[abs(result1[i]-resultr[i])/resultr[i]*100 for i in range(1,len(resultr))],label='percentage error 200 PN')
plt.plot(range(2,100),[abs(result2[i]-resultr[i])/resultr[i]*100 for i in range(1,len(resultr))],label='percentage error 60 PN')
plt.ylabel('precession percentage error')
plt.xlabel('mode')
#plt.ylim([0,100])
```
